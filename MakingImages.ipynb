```python
import cv2
from cv2 import imread, imwrite, resize
import uuid
import os
import time
```

```python
labels = ['ThumbsUp','ThumbsDown']
numimg = 100
```

```python
paths = {
    "Workspace": os.path.join('Workspace'),
    "GoodOrBadModel" : os.path.join('Workspace','GoodOrBad'),
    "Images" : os.path.join('Workspace','GoodOrBad','Images'),
    "CollectedImages" : os.path.join('Workspace','GoodOrBad','Images','CollectedImages'),
    "ProcessedImages" : os.path.join('Workspace','GoodOrBad','Images',"ProcessedImages"),
    "Annotations" : os.path.join('Workspace','GoodOrBad','Annotations')
    
}
```

```python
for path in paths:
    if not os.path.exists(paths[path]):
        os.makedirs(paths[path])
```

check to see if cap works

```python
while True:
    cap = cv2.VideoCapture(0)
    cv2.startWindowThread()
    ret, frame = cap.read()
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
cap.release()
cv2.waitKey(1)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

make the images

```python
for label in labels:
    cap = cv2.VideoCapture(0)
    cv2.startWindowThread()    
    print('collecting images for {}'.format(label))
    time.sleep(5)
    for imgnum in range(numimg):
        print('collecting image number{}'.format(imgnum))
        
        ret, frame = cap.read()
        cv2.waitKey(1)
        cv2.imshow('frame',frame)
        imagesavepath = os.path.join(paths['CollectedImages'],label + '.'+'{}.jpg'.format(str(uuid.uuid1())))
        print(imagesavepath)
        cv2.waitKey(1)
        time.sleep(1)
        cv2.imwrite(imagesavepath,frame)
        cv2.waitKey(1)
        
    if cv2.waitKey(1) &0xFF == ord('q'):
        break
cap.release()
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
#making 1 off image
cap = cv2.VideoCapture(0)
cv2.startWindowThread()    
print("ThumbsDown")
time.sleep(5)
ret, frame = cap.read()
cv2.waitKey(1)
cv2.imshow('frame',frame)
name = "ThumbsDown" + '.'+'{}.jpg'.format(str(uuid.uuid1()))
imagesavepath = os.path.join(paths['CollectedImages'],name)
cv2.imwrite(imagesavepath,frame)
cap.release()
cv2.waitKey(1)
cv2.destroyAllWindows()
img = imread(imagesavepath)
ResizedImage = resize(img, (640,640))
SavePath = os.path.join(paths["ProcessedImages"],name)
imwrite(SavePath,ResizedImage)
```

resize and move the images

```python

images = []
for image in os.listdir(paths['CollectedImages']):
    images.append(image)
for image in images:
    img = imread(os.path.join(paths['CollectedImages'],image))
    ResizedImage = resize(img, (640,640))
    SavePath = os.path.join(paths["ProcessedImages"],image)
    imwrite(SavePath,ResizedImage)
    
```
